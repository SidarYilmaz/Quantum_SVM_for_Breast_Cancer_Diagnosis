```python
import sys
sys.path.append(r"c:\Users\Administrator\Desktop\Quantum_SVM_for_Breast_Cancer_Diagnosis\src")

from models.quantum_kernel import QuantumKernel
from models.qsvm_model import QSVM
import numpy as np

```

```python

# Preprocessing çıkışları
X_train = np.load("../data/processed/X_train_qsvm.npy")
X_test  = np.load("../data/processed/X_test_qsvm.npy")
y_train = np.load("../data/processed/y_train.npy")
y_test  = np.load("../data/processed/y_test.npy")
```

```python
n_features = X_train.shape[1]

qkernel = QuantumKernel(n_features)
qsvm = QSVM(qkernel)
```

```python

qsvm.fit(X_train, y_train)
acc = qsvm.accuracy(X_test, y_test, X_train)

print("QSVM (Qiskit amplitude kernel) accuracy:", acc)
```
